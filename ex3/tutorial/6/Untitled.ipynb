```python
import numpy as np
from scipy import special
import matplotlib
import matplotlib.pyplot as plt

print(matplotlib.matplotlib_fname())
params = {
   'axes.labelsize': 8,
   'font.size': 8,
   'legend.fontsize': 9,
   'xtick.labelsize': 9,
   'ytick.labelsize': 9,
   'text.usetex': False,
   'figure.figsize': [8, 5],
   'errorbar.capsize': 3,
   'lines.linewidth': .8,
   'font.family': 'CMU Serif',
   'mathtext.fontset': 'cm'
}
plt.rcParams.update(params)
```

```python
x = np.linspace(0, 1, 100)
s = ["-", "--", ":", "-."]
for n in [1,2,3]:
    plt.plot(x, np.sqrt(2) * np.sin(np.pi*x*n)**2, 
         label="$n_x = {}$".format(n), color="black", linestyle=s[n - 1])
plt.legend()
plt.xlabel("x/d")
plt.ylabel(r"$|{\langle x | n_x \rangle}|^2$")
plt.savefig("1.pdf")
plt.show()
```

```python
n = 50
a = 1
b = 2
x = np.linspace(0, 3, 50)
y = np.linspace(0, 1, 50) 
z = np.array([np.sqrt(4/3) * np.sin(np.pi*i/3*b)**2 * np.sin(np.pi*j*a)**2 for j in y for i in x])
Z = z.reshape(50, 50)
plt.imshow(Z, interpolation='bilinear', cmap='gray_r', extent=[0, 3, 0, 1])
plt.xlabel("y/d")
plt.ylabel("x/d")
plt.savefig("2.pdf")
plt.show()
```

```python
n = 200
a = 5
b = 3
x = np.linspace(0, 3, n)
y = np.linspace(0, 1, n) 
z = np.array([np.sqrt(4/3) * np.sin(np.pi*i/3*b)**2 * np.sin(np.pi*j*a)**2 for j in y for i in x])
Z = z.reshape(n, n)
plt.imshow(Z, interpolation='bilinear', cmap='gray_r', extent=[0, 3, 0, 1])
plt.xlabel("y/d")
plt.ylabel("x/d")
plt.savefig("3.pdf")
plt.show()
```

```python
x = np.linspace(-4, 4, 1000)
s = ["-", "--", ":", "-."]
for n in [0, 1, 2]:
    plt.plot(x, np.abs(np.sqrt(1/(np.sqrt(np.pi) * 2**n * np.math.factorial(n))) * special.hermite(n)(x)*np.exp(-(x**2)/2))**2, 
         label="$n = {}$".format(n), color="black", linestyle=s[n])
plt.legend()
plt.xlabel("x/d")
plt.ylabel(r"$|\phi_n|^2$")
plt.savefig("4.pdf")
plt.show()
```
